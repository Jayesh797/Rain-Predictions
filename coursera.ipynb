```python
import pandas as pd
df=pd.read_csv('Weather_Data.csv')

transpose=df.T
transpose
```

```python
df
```

```python
df_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

```python
df_processed
```

```python
df_processed.drop(columns=['Date'],inplace=True)
```

```python
df_processed
```

```python
df_processed.replace({False: 0, True: 1}, inplace=True)
```

```python
df_processed
```

```python
X = df_processed.drop(columns='RainTomorrow', axis=1)
Y = df_processed['RainTomorrow']
```

```python
X
```

```python
Y
```

```python
Y.replace({'Yes':1,'No':0},inplace=True)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
y_train
```

1)Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

model=LinearRegression();
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("Mean absolute Error:", mae)
print("R2 Score:", r2)
print("Mean Squared Error:", mse)
```

```python
metrics_df = pd.DataFrame({
    'Metric': ['Mean Absolute Error (MAE)', 'Mean Squared Error (MSE)', 'R-squared (R2)'],
    'Value': [mae, mse, r2]
})

metrics_df
```

2)KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,jaccard_score,f1_score

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
```

```python
KNN_Accuracy_Score = accuracy_score(y_test,y_pred)
KNN_JaccardIndex = jaccard_score(y_test,y_pred,average='weighted')
KNN_F1_Score = f1_score(y_test,y_pred,average='weighted')
```

```python
metrics=pd.DataFrame({
    'Metrics':['KNN_Accuracy_Score','KNN_JaccardIndex','KNN_F1_Score'],'value':[KNN_Accuracy_Score,KNN_JaccardIndex,KNN_F1_Score]
})
metrics
```

3)Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,jaccard_score,f1_score
```

```python
tree=DecisionTreeClassifier()
tree.fit(x_train,y_train)
y_pred=tree.predict(x_test)

Tree_Accuracy_Score = accuracy_score(y_test,y_pred)
Tree_JaccardIndex = jaccard_score(y_test,y_pred,average='weighted')
Tree_F1_Score = f1_score(y_test,y_pred,average='weighted')
```

```python
metrics=pd.DataFrame({
    'Metrics':['tree_Accuracy_Score','tree_JaccardIndex','tree_F1_Score'],'value':[Tree_Accuracy_Score,Tree_JaccardIndex,Tree_F1_Score]
})
metrics
```

4)Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, jaccard_score, f1_score, log_loss
LR=LogisticRegression(max_iter=1000)
LR.fit(x_train,y_train)
y_predictions=LR.predict(x_test)
```

```python
y_predict_proba=LR.predict_proba(x_test)
```

```python
LR_Accuracy_Score = accuracy_score(y_test,y_predictions)
LR_JaccardIndex = jaccard_score(y_test,y_predictions,average='weighted')
LR_F1_Score = f1_score(y_test,y_predictions,average='weighted')
LR_Log_Loss = log_loss(y_test,y_predict_proba)
```

```python
metrics=pd.DataFrame({
    'Metrics':['LR_Accuracy_Score','LR_JaccardIndex','LR_F1_Score','LR_Log_Loss'],'value':[LR_Accuracy_Score,LR_JaccardIndex,LR_F1_Score,LR_Log_Loss]
})
metrics
```

5) SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,jaccard_score,f1_score

svm=SVC()
svm.fit(x_train,y_train)
y_pred=svm.predict(x_test)
```

```python
SVM_Accuracy_Score = accuracy_score(y_test,y_pred)
SVM_JaccardIndex = jaccard_score(y_test,y_pred,average='weighted')
SVM_F1_Score = f1_score(y_test,y_pred,average='weighted')
```

```python
metrics=pd.DataFrame({
    'Metrics':['SVM_Accuracy_Score','SVM_JaccardIndex','SVM_F1_Score'],'value':[SVM_Accuracy_Score,SVM_JaccardIndex,SVM_F1_Score]
})
metrics
```
